```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0 , 2 ,10)
y= x**2
z= y**3
```

**1. Construire une courbe avec x en abscisse et y en ordonnée**

```python
plt.plot(x,y)
plt.xlabel('Axe x')
plt.ylabel('Axe y')
```

**2. Changer l'épaisseur et la couleur de la ligne du graphique précédemment créé
avec le paramètre "c" et "lw"**

```python

plt.plot(x , y ,c='pink',lw=5.5 ) #c=color , lw=lineWidth
plt.xlabel('Axe x')
plt.ylabel('Axe y')
```

**3. Construire sur le même graphique deux courbes :**

```python
#1 Une première courbe avec x en abscisse et y en ordonnée
plt.plot(x,y,label='xy')

#2 Une deuxième courbe avec x en abscisse et z en ordonnée
plt.plot(x,z,label='xz')

#3 Ajouter une légende au graphique
plt.legend()

#4 Ajouter un titre au graphique
plt.title('Representation des courbes xy et xz')

#5 Ajouter un titre à chaque axe
plt.xlabel('Axe x')
plt.ylabel('Axe y et z')

plt.show();
```

**11. Stockez dans "liste qual" les noms des variables qualitatives et dans “liste_quant”
celles des variables quantitatives.**

```python
list_qual=[]
list_quant=[]

for index in df.columns :
    if df[index].dtypes== 'O' :
        list_qual.append(index)
    else:
        list_quant.append(index)
print('La liste des var qualitatives est : ',list_qual)
print('La liste des var quantitatives est : ',list_quant)
```

**12. En utilisant “liste_quant”, faites une représentation graphique de chaque variable
quantitative en utilisant une boucle for.**

```python
plt.figure(figsize=[17,6])

plt.gcf().subplots_adjust( hspace = 0.5) #hspace = 0.2 : espacement vertical entre deux graphiques(valeur par defaut)

i=1
for var in list_quant :
    plt.subplot(2,4,i)
    plt.hist(df[var],edgecolor='white')
    plt.title(var)
    i+=1
```

**13. En utilisant “liste_qual”, faites une représentation graphique de chaque variable

```python
plt.figure(figsize=[17,6])

plt.gcf().subplots_adjust( hspace = 0.5) #hspace = 0.2 : espacement vertical entre deux graphiques(valeur par defaut)

i=1
for var in list_qual :
    plt.subplot(2,3,i)
    x= df[var].value_counts()
    x.plot(kind='bar')
    plt.title(var)
    i+=1
```
