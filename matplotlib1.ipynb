```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Construire une courbe avec x en abscisse et y en ordonnée

x = np.linspace(0, 2, 10)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0, 2, 10)
y = x**2
plt.plot(x,y, color="r", linewidth=4)
plt.show()
```

```python
#3. Construire sur le même graphique deux courbes :
#1. Une première courbe avec x en abscisse et y en ordonnée
#2. Une deuxième courbe avec x en abscisse et z en ordonnée
#3. Ajouter une légende au graphique
#4. Ajouter un titre au graphique
#5. Ajouter un titre à chaque axe

x = np.linspace(0, 2, 10)
y = x**2
z = x**3

plt.plot(x,y, color="r", linewidth=3, label="courge1")
plt.plot(x,z, label="courbe2")
plt.legend()
plt.xlabel("abscisses")
plt.ylabel("ordonnées")
plt.show()
```

```python
#Afficher à travers un boxplot la distribution de x
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

x = np.linspace(0, 2, 10)
y = x**2
z = x**3
plt.plot(x,y, color="r", linewidth=3, label="courge1")
plt.plot(x,z, label="courbe2")
plt.legend()
plt.xlabel("abscisses")
plt.ylabel("ordonnées")
plt.boxplot(x)
plt.show()
```
