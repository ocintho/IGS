```python
#Calculer la médiane et la moyenne des colonnes
print("mediane age",dataset["Age"].median())
print("mediane Customer spendings",dataset["Customer spendings"].median())

print("moyenne age",dataset["Age"].mean())
print("moyenne Customer spendings",dataset["Customer spendings"].mean())

```

```python
#recoder l’age en variable qualitative ne faisant des inter

def categories_age(age):
    if age < 18:
        return 'mineur'
    elif age < 38:
        return 'jeune'
    elif age < 68:
        return 'adulte'
    else:
        return 'vieux'

dataset['categorie age'] = dataset['Age'].map(categories_age)
dataset.drop(['Age'], axis=1, inplace=True)
```

```python
#Nettoyer les lignes avec moins de 10 € de dépenses par client : supprimez toutes les
#lignes d’utilisateurs ayant dépensé moins de 10 € sur le site,

dataset.drop( dataset[ (dataset['Customer spendings'] < 10)].index, inplace=True)
dataset
```

```python
#Nettoyer les doublons : supprimer toutes les lignes qui apparaissent plus d’une fois
#dans les données,
dataset.drop_duplicates( keep = 'first' , inplace = True , ignore_index = False )
```
